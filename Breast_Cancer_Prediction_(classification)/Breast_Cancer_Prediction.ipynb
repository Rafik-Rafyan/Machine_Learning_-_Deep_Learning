### Checking Outliers

```python
# Lets create a function wich will show us diagnostic of data with two ways.
def vis_outliers(df, variable):
    plt.figure(figsize=(14,20))
    plt.title(variable)

    # histogram
    plt.subplot(3,1,1)   # tree rows and one column place, where histogram is the first visualisation
    plt.hist(df[variable])
    # boxplot
    plt.subplot(3,1,2)   # tree rows and one column place, where boxplot is the second visualisation
    sns.boxplot(x = df[variable])
    # Lets see data values as points
    plt.subplot(3,1,3)  #  tree rows and one column place, where scatterplot is the third visualisation
    sns.scatterplot(df[variable])
    
```

```python
import warnings
warnings.filterwarnings("ignore")

for i in x_train.columns:
    vis_outliers(x_train, i)
```

```python
# We have low outliers for all predictors, so I want to check maybe those are not outliers but indicators of cancer's presence.
# If thos are indicators of Cancer presence so in image below count of 0 must be significantly less than 1․
sns.countplot(data=x_train, x='Cancer')
```

```python
# As we saw those are outliers: they don't indicate to cancer's presence, so I choose to delete them using IQR method.
    
for i in x_train.drop("Cancer", axis=1).columns:
    Q1 = x_train[i].quantile(0.25)
    Q3 = x_train[i].quantile(0.75)
    IQR = Q3 - Q1
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q3 - 1.5 * IQR
    x_train = x_train.loc[(x_train[i] > lower_limit) & (x_train[i] < upper_limit)]
    
```

```python
x_train.corr()["Cancer"]
```

### Separating x_train into X and y

```python
X = x_train.drop("Cancer", axis = 1)
y = x_train["Cancer"]
```

# Building Model

```python
# No need for train_test_split, because data already splitted
```

```python
# Creating model
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()  
```

```python
# Normalizing data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Using Pipeline to automatically normalize new data for prediction.
from sklearn.pipeline import Pipeline
pipe = Pipeline([("scaler", scaler), ("Knn", model)])
```

```python
# As I want to check which K-value is better from 1 to 30, I use GridSearchCV.
from sklearn.model_selection import GridSearchCV
full_model = GridSearchCV(pipe, param_grid={"Knn__n_neighbors":list(range(1,30,2))}, cv=2, verbose=2, scoring="accuracy")
```

```python
full_model.fit(X, y)                    # Fitting model
```

```python
y_pred = full_model.predict(x_test)     # Predict from x_test data
```

```python
full_model.best_params_  # Let's see which K-value was more optimal
```
